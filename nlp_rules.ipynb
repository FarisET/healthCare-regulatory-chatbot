```python
import spacy

nlp = spacy.load("en_core_web_md")
with open("data/jci_standards.txt", "r", encoding="utf-8", errors="replace") as f:
    text = f.read()

doc = nlp(text)
```

```python
# This cell is labelling CLAUSES. By default, spacy md model was labelling all clauses as ORG

#nlp = spacy.load("en_core_web_md")  # or whichever model
# ruler = EntityRuler(nlp, overwrite_ents=True)  # overwrite stats NER if needed

# token pattern:
#  - a 3-letter code + dot + number + optional .number (use REGEX on TEXT)
#  - then 1+ tokens (the clause)
pattern = [
    {"TEXT": {"REGEX": r"[A-Z]{3}\.\d+(?:\.\d+)*"}},  # e.g. IPSG.2 or IPSG.2.1 or IPSG.2.1.3
    {"IS_SPACE": True, "OP": "*"},
    {"IS_ALPHA": True, "OP": "+"}                     # the clause: one or more alpha tokens
]

# ruler.add_patterns([{"label": "CLAUSE", "pattern": pattern}])
ruler = nlp.add_pipe("entity_ruler", before="ner") # insert before NER so it takes precedence
ruler.add_patterns([{"label": "CLAUSE", "pattern": pattern}])
doc = nlp(text)
for ent in doc.ents:
    print(ent.label_, ent.text)
```
